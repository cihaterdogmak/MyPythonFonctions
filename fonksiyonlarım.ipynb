```python
# String bir metni istenilen karakterlerden ayırıp bir listeye tanımlama .split() işlevli


def ayırstr(string, ayırıcıparam = " ") :
    liste = []
    eklenecekparam = ""
    string = string + ayırıcıparam
    print(string)
    for i in string :
        if i == ayırıcıparam :
            liste += [eklenecekparam]
            eklenecekparam = ""
        else :
            eklenecekparam += i
    
    return liste
```

```python
# Listedeki elemanları sıralayan fonksiyon


def sirala(list) :
    for i in range(len(list)):
        for j in range(len(list)-i-1) :
            if list[j] > list[j+1] :
                list[j],list[j+1] = list[j+1],list[j]
    return list
```

```python
# String listeyi İntager listeye çeviren fonksiyon


def listint(list) :
    intlist = []
    for i in list :
        intlist += [int(i)]
    return intlist
```

```python
# Float bir sayıyı tavana yuvarlama math.ceil()


def ceil(x):
    z = "float"
    tmp = ""
    
    if x == (x // 1) :
        z = "intager"

    x = str(x)
    for i in x :
        if i == "." :
            if z == "float" :
                return int(tmp) + 1 
            elif z == "intager" :
                return int(tmp)
        else :
            tmp += i
    return int(tmp)
    

x = 19.00

print(ceil(x))
```

```python
# Float bir sayıyı tabana yuvarlama math.floor()


def floor(x):
    z = "float"
    tmp = ""
    
    if x == (x // 1) :
        z = "intager"

    x = str(x)
    for i in x :
        if i == "." :
            return int(tmp)
        else :
            tmp += i
    return int(tmp)
```

```python
# Binary search
# Gerekli fonksiyonlar  sirala(), floor()

def binarysearch(arr, key, low = 0, high = 0):
    if high == 0 :
        high = len(arr)-1

    arr = sirala(arr)
    print(arr)  #dizeyi sıralıyken yazdıran kısım
    
    while(low <= high) :
        mid = floor((low + high) / 2)
        if arr[mid] < key :
            low = mid + 1
        elif arr[mid] > key :
            high = mid - 1
        else :
            return mid
    return -1   # -1 ise dizede yoktur
```
